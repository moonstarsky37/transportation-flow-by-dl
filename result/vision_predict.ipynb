```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
predict = np.load('STresnet_prediction.npy')
groundtruth = np.load('STresnet_groundtruth.npy')
predict.max(), predict.shape, groundtruth.max(), groundtruth.shape
```

```python
np.random.seed(0)
index = np.random.randint(predict.shape[0])
index
```

```python
predict_data=(predict[index]/2+0.5)*45
groundtruth_data = (groundtruth[index]/2+0.5)*45
inflow_predict = predict_data[0]
inflow_groundtruth = groundtruth_data[0]
```

```python
%%time

#annot(矩阵上数字),annot_kws(矩阵上数字的大小颜色字体)matplotlib包text类下的字体设置

%matplotlib inline 
 
f, (ax1, ax2, ax3) = plt.subplots(figsize=(30,8),ncols=3)
ax1.set_title('Prediction')
sns.heatmap(inflow_predict, annot=True, ax=ax1, cmap="cool")
ax2.set_title('Target')
sns.heatmap(inflow_groundtruth, annot=True, ax=ax2, cmap="cool")
ax3.set_title(f'Target - prediction , and sum of ABS(Target-Pred)={np.sum(np.abs(inflow_groundtruth-inflow_predict)):.3f}')
sns.heatmap(np.round(inflow_groundtruth-inflow_predict,1), annot=True, ax=ax3, cmap="coolwarm", vmin=-5, vmax=5,)
# plt.savefig(f"./vis_results/img_{index:>04}.jpg", dpi=100, bbox_inches='tight', pad_inches=0)
# plt.close()
# Keyword arguments for ax.text when annot is True.  http://stackoverflow.com/questions/35024475/seaborn-heatmap-key-words
```

```python
%%time
for index in range(len(predict[:5])):
    predict_data=(predict[index]/2+0.5)*45
    groundtruth_data = (groundtruth[index]/2+0.5)*45

    inflow_predict = predict_data[0]
    inflow_groundtruth = groundtruth_data[0]

    f, (ax1, ax2, ax3) = plt.subplots(figsize=(30,8),ncols=3)
    ax1.set_title('Prediction')
    sns.heatmap(inflow_predict, annot=True, ax=ax1, cmap="cool")
    ax2.set_title('Target')
    sns.heatmap(inflow_groundtruth, annot=True, ax=ax2, cmap="cool")
    ax3.set_title(f'Target - prediction , and sum of ABS(Target-Pred)={np.sum(np.abs(inflow_groundtruth-inflow_predict)):.3f}')
    sns.heatmap(np.round(inflow_groundtruth-inflow_predict,1), annot=True, ax=ax3, cmap="coolwarm", vmin=-5, vmax=5,)
    plt.savefig(f"./vis_results/img_{index:>04}.jpg", dpi=100, bbox_inches='tight', pad_inches=0)
    plt.close()
    print(f"Process img_{index:>04}.jpg done!")
```
